```python
from random import randrange

def searchLandingArea(img_edges, n_searches=1000, kernel_size=20, n_kernel_samples=80, treshold=0.8):
    shape = img_edges.shape
    landing_points = []
    for i in range(n_searches):
        sample_x = randrange(kernel_size//2,shape[0]-kernel_size//2)
        sample_y = randrange(kernel_size//2,shape[1]-kernel_size//2)
        fitness = 0
        fitness = img_edges[sample_x-kernel_size//2:sample_x+kernel_size//2,
                          sample_y-kernel_size//2:sample_y+kernel_size//2].sum()
        if 1.0 - fitness/(kernel_size*kernel_size*255.0) > treshold:
            landing_points.append([sample_x,sample_y])
    return np.array(landing_points)
```

```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt

img = cv.imread('city.jpg',0)
edges = cv.Canny(img,100,200)

f = plt.figure(figsize=(15,15))
ax = f.add_subplot(121)
ax2 = f.add_subplot(122)
ax.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
ax2.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])

landing_points = searchLandingArea(edges,1000)

ax2.plot(landing_points[:,1],landing_points[:,0], 'bo')

plt.show()
```
