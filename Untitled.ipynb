```python
from laspy.file import File
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
inFile = File(fr'161104_hq_group1_densified_point_cloud.las', mode='r')
```

```python
print(inFile.header.scale)
```

```python
rgb = np.array([inFile.Red,inFile.Green,inFile.Blue])
```

```python
xyz = np.array((inFile.x,inFile.y,inFile.z))
```

```python
print(f'max rgb = {np.max(rgb[0])},{np.max(rgb[1])},{np.max(rgb[2])} || min rgb = {np.min(rgb[0])},{np.min(rgb[1])},{np.min(rgb[2])}')
```

```python
# normalize rgb
rgb[0] = rgb[0]/np.max(rgb[0]) * 255.0
rgb[1] = rgb[1]/np.max(rgb[1]) * 255.0
rgb[2] = rgb[2]/np.max(rgb[2]) * 255.0
```

```python
rgb.shape
```

```python
print(f'max rgb = {np.max(rgb[0])},{np.max(rgb[1])},{np.max(rgb[2])} || min rgb = {np.min(rgb[0])},{np.min(rgb[1])},{np.min(rgb[2])}')
```

```python
# XYZ normalizing and rastering
```

```python
print(f'max xyz = {np.max(xyz[0])},{np.max(xyz[1])},{np.max(xyz[2])} || min xyz = {np.min(xyz[0])},{np.min(xyz[1])},{np.min(xyz[2])}')
```

```python
# Shift to min val = 0.0
xyz[0] = xyz[0] - np.min(xyz[0])
xyz[1] = xyz[1] - np.min(xyz[1])
xyz[2] = xyz[2] - np.min(xyz[2])
```

```python
print(f'max xyz = {np.max(xyz[0])},{np.max(xyz[1])},{np.max(xyz[2])} || min xyz = {np.min(xyz[0])},{np.min(xyz[1])},{np.min(xyz[2])}')
```

```python
X = (xyz[0] * 100).astype(int)
Y = (xyz[1] * 100).astype(int)
Z = np.int32(xyz[2])
# C = np.array([[255, 0, 0], [0, 255, 0], [0, 0, 255]])
fig = plt.figure(figsize=(20,20))
ax = fig.add_subplot(111)

ax.scatter(X, Y, c = Z)
plt.show()
```
